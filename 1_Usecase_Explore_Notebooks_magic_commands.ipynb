Task6: Create a pyspark dataframe df1 reading the data from the above file using pyspark magic command %python


Task7: Write the above dataframe df1 data into a databricks table called 'default.mobile_os_usage' using pyspark magic command %python



```python
df_clean=(df
          .withColumnRenamed("Mobile Operating System","Mobile_Operating_System")
          .withColumnRenamed("Percent of Usage","Percent_of_Usage"))

df_clean.write.mode("overwrite").saveAsTable("default.mobile_os_usage")
```
